```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
from sklearn.datasets import load_diabetes
```

```python
data=load_diabetes()
```

```python
mdata=data.data
```

```python
tdata=data.target
```

```python
fdata=pd.DataFrame(mdata,columns=data.feature_names)
```

```python
fdata
```

```python
tdata=pd.DataFrame(tdata,columns=['target'])
```

```python
tdata
```

```python
fdata=pd.concat([fdata,tdata],axis=1)
fdata=fdata.iloc[:,[2,10]].values
```

```python
fdata
```

```python
from sklearn.cluster import KMeans
```

```python
elb=[]
for i in range(1,20):
    km=KMeans(n_clusters=i)
    km.fit(fdata)
    elb.append(km.inertia_)
```

```python
elb
```

```python
plt.plot(range(1,20),elb)
plt.scatter(range(1,20),elb)
```

```python
fkm=KMeans(n_clusters=4)
```

```python
fkm.fit(fdata)
```

```python
pr=fkm.predict(fdata)
```

```python
pr
```

```python
plt.scatter(fdata[pr==0,0],fdata[pr==0,0],label='kanjoos')
plt.scatter(fdata[pr==1,0],fdata[pr==1,0],label='okok')
plt.scatter(fdata[pr==2,0],fdata[pr==2,0],label='dildar')
plt.scatter(fdata[pr==3,0],fdata[pr==3,0],label='loota hoon')
plt.scatter(fkm.cluster_centers_[:,0],fkm.cluster_centers_[:,0],label="Cluster",s=100)
plt.legend()
```
